# **Importação**


Importação da pasta 'Vendas' direto do Google Drive.

Importação de Pandas para tratar os arquivos.

Importação do Plotly para gerar gráficos.

```python
import pandas as pd

import plotly.express as px

import os

lista_arquivo = os.listdir('/content/drive/MyDrive/projeto python/Vendas')
print(lista_arquivo)
```

# **Condição**

**FOR** para listar os arquivos dentro da 'lista_arquivo'.

**IF** como condição para arquivos de venda.

```python
for arquivo in lista_arquivo:

  if 'Vendas' in arquivo:
    # print(f'/content/drive/MyDrive/projeto python/Vendas/{arquivo}') para ver o caminho
    print(arquivo)

  else:
    print('Arquivo de Devolução')
```

# **Tratamento de dados**


Compilação de tabelas com **CONCAT**.

```python
tabela_total = pd.DataFrame()

for arquivo in lista_arquivo:
    if 'Vendas' in arquivo:
        tabela = pd.read_csv(f'/content/drive/MyDrive/projeto python/Vendas/{arquivo}')
        tabela_total = pd.concat([tabela_total, tabela], ignore_index=True)

display(tabela_total)
```

# **Cálculos**

Produto mais vendido

```python
tabela_produtos = tabela_total.groupby('Produto').sum()

display(tabela_produtos[['Quantidade Vendida', 'Preco Unitario']].sort_values(by='Quantidade Vendida', ascending=False))
```

Produto que mais faturou

```python
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']

tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by='Faturamento', ascending=False)

display(tabela_faturamento)
```

Loja/Cidade com maior faturamento

```python
tabela_lojas = tabela_total.groupby('Loja').sum()

display(tabela_lojas['Faturamento'])
```

# Gráfico


Criação de gráfico de barras.

```python
grafico = px.bar(tabela_lojas, x= tabela_lojas.index, y= 'Faturamento')

grafico.show()
```
